Object olan değişkenlerde, en çok tekrar eden değerle null değerleri doldurabiliriz. Veya 2. bir yöntem olarak kendisinden bir önceki değerle null değeri doldurabiliriz eğer veriler sıralıysa.

Object olan verilerde şu şekilde yapabiliriz;


```python
df.STATE.value_counts()
```

Buradaki çoğu değişken 33 defa kullanılmış.O zaman null olan değeri kendisinden bir önceki değerle doldurabiliriz.(Sıralıysa) Eksik değerlerin doldurulması veri kaybını önlemek ve yanlış analiz yapmanın önüne geçmek için önemlidir.

```python
new_data = df[df['YEAR']==1992]        #1992 yılındaki değerleri ele alalım.
new_data
```

TOTAL_REVENUE ile çalışalım. Buradaki NaN değerleri mean ile dolduralım.


```python
import warnings

warnings.filterwarnings('ignore')
new_data["TOTAL_REVENUE"].fillna(new_data["TOTAL_REVENUE"].mean(), inplace=True)
```

```python
new_data
```

```python
new_data.TOTAL_REVENUE.isnull()
```

```python
new_data.FEDERAL_REVENUE.interpolate(method ='linear',limit_direction ='forward', inplace=True)
```

```python
new_data
```

```python
new_data.FEDERAL_REVENUE.isnull()
```

```python
new_data.isnull().sum()*100/df.shape[0]   #total ve federal revenue'nun null değerlerinin sıfırlandığını görebiliriz.
```
